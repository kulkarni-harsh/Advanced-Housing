```python
## import libs
import pandas as pd 
import numpy as np
```

```python
## take dataset
dataset_og=pd.read_csv('train.csv')
dataset=dataset_og.copy()
dataset
```

```python
## lets focus on Categorical Null Values
features_cat_na=[feature for feature in dataset.columns if dataset[feature].dtype=='O'and dataset[feature].isnull().sum()>1]
features_cat_na
for feature in features_cat_na:print(f"{feature} has {np.round(dataset[feature].isnull().mean(),4)}")
```

```python
def replace_cat_features(dataset,features_with_na):
    data=dataset.copy()
    data[features_with_na]=data[features_with_na].fillna('Missing')
    return data
```

```python
dataset=replace_cat_features(dataset,features_cat_na)
dataset[features_cat_na].isnull().sum()
```

```python
dataset.head()
```

```python
features_num_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
```

```python
for feature in features_num_na:
    print("{}: {}% missing value".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in features_num_na:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```

```python
dataset[features_num_na].isnull().sum()
```

```python
dataset.head(10)
```

```python
## temporal variables
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
log_num_features=[feature for feature in features_num if dataset[feature].min()>0 and feature not in ['Id']and 'Yr'not in feature and 'Year' not in feature]
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```
